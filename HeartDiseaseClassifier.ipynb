```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('processed.cleveland.data',header=None)
```

      4. restbps - resting blood pressure (in mm Hg) 
      5. chol - cholesterol

```python
df.columns=['age',
            'sex',
            'cp',
            'restbps',
            'chol',
            'fbs',
            'restecg',
            'thalach',
            'exang',
            'oldpeak',
            'slope',
            'ca',
            'thal',
            'hd']
```

```python
len(df.loc[(df['ca']=='?') | (df['thal']=='?')])
```

```python
df.loc[(df['ca']=='?') | (df['thal']=='?')]
```

```python
df_new=df.loc[(df['ca']!='?') & (df['thal']!='?')]
```

```python
X=df_new.drop('hd',axis=1).copy()
X.head()
```

```python
y=df_new['hd'].copy()
```

```python
y_not_zero=y>0
#get the index for each non zero value in y
y[y_not_zero]=1 
#set them to 1
y.unique()
```

age: age in years, FLOAT

sex: sex (1 = male; 0 = female), CATEGORICAL

cp: chest pain type, CATEGORICAL

   -- Value 1: typical angina
   -- Value 2: atypical angina
   -- Value 3: non-anginal pain
   -- Value 4: asymptomatic 

restbps: resting blood pressure (in mm Hg), FLOAT

chol: serum cholestoral in mg/dl, FLOAT

fbs: (fasting blood sugar > 120 mg/dl)  (1 = true; 0 = false), CATEGORICAL

restecg: resting electrocardiographic results, CATEGORICAL

   -- Value 0: normal
   -- Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
   -- Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria

thalach: maximum heart rate achieved, FLOAT

exang: exercise induced angina (1 = yes; 0 = no), CATEGORICAL

oldpeak = ST depression induced by exercise relative to rest, FLOAT

slope: the slope of the peak exercise ST segment, CATEGORICAL

   -- Value 1: upsloping
   -- Value 2: flat
   -- Value 3: downsloping

ca: number of major vessels (0-3) colored by flourosopy, FLOAT

thal: thallium heart scan, CATEGORICAL

   -- 3 = normal
   -- 6 = fixed defect
   -- 7 = reversable defect


```python
X_encoded=pd.get_dummies(X,columns=['cp','restecg','slope','thal'],dtype='float64')
#default dtype is bool
X_encoded.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_encoded,y,random_state=42)
```

```python

scaler = StandardScaler()
cols_to_scale = ['age', 'restbps', 'chol', 'thalach', 'oldpeak', 'ca']  

X_train[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])
X_test[cols_to_scale] = scaler.transform(X_test[cols_to_scale])
```

```python
X_train[:5]
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=22))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='sgd',metrics=['accuracy'])

```

```python
history = model.fit(
    X_train,
    y_train,
    batch_size=32,
    epochs=100,
    # We pass some validation for
    # monitoring validation loss and metrics
    # at the end of each epoch
    validation_data=(X_test, y_test),
)
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_acc:.4f}")
```
