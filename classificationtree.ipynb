There are two methods to perform one hot encoding, one by using ColumnTransformer from scikitlearn, and another by using get_dummies from pandas. Both have their own pros and cons.

```python
X_encoded=pd.get_dummies(X,columns=['cp','restecg','slope','thal'],dtype='float64')
#default dtype is bool
X_encoded.head()
```

The independent variable 'hd' should contain 5 different levels of heart disease, with 0=no heart disease, and 1-4 are various degrees of heart disease.
In this notebook we are going to perform binary classification, so we need to convert all values >0 to 1.

```python
y.unique()
```

```python
y_not_zero=y>0
#get the index for each non zero value in y
y[y_not_zero]=1 
#set them to 1
y.unique()
```

# Building a preliminary tree

```python
X_train,X_test,y_train,y_test=train_test_split(X_encoded,y,random_state=42)

clf_dt=DecisionTreeClassifier(random_state=42)
clf_dt=clf_dt.fit(X_train,y_train)
```

```python
plt.figure(figsize=(15,7.5))
plot_tree(clf_dt,filled=True,rounded=True,class_names=['No HD','YES HD'],feature_names=X_encoded.columns)
```

```python
ConfusionMatrixDisplay.from_estimator(
    clf_dt, X_test, y_test)
```

In the Confusio matrix, it is seen that of the 31+11=42 people that did not have Heart Disease, 31 (74%) were correctly classified. And of the 7+26=33 people that have Heart Disease, 26 (79%) were correctly classified.

Let's try to improve this performance by pruning the decision tree.

# Cost Complexity Pruning 

## Part-1: Visualize alpha

Let's extract the different alpha values that are available for this tree and build a pruned tree for each value for alpha.

NOTE: we omit the maximum value for alpha because it would prune all the leaves, leaving us with only a root instead of a tree.

```python
path=clf_dt.cost_complexity_pruning_path(X_train,y_train)
ccp_alphas=path.ccp_alphas
ccp_alphas=ccp_alphas[:-1]

clf_dts=[]

for ccp_alpha in ccp_alphas:
    clf_dt=DecisionTreeClassifier(random_state=0,ccp_alpha=ccp_alpha)
    clf_dt.fit(X_train,y_train)
    clf_dts.append(clf_dt)
    
```

```python
train_scores=[clf_dt.score(X_train,y_train) for clf_dt in clf_dts]
test_scores=[clf_dt.score(X_test,y_test) for clf_dt in clf_dts]

fig,ax=plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs Alpha for training and testing sets")
ax.plot(ccp_alphas,train_scores,marker='o',label='train',drawstyle='steps-post')
ax.plot(ccp_alphas,test_scores,marker='o',label='test',drawstyle='steps-post')
ax.legend()
plt.show()
```

In the graph above, we see that accuracy for the testing dataset hits maximum value when alpha is about 0.016.



Since there are many ways we could have divided the original dataset into Training and Testing datasets, different training and testing datasets might suggest a different optimal value for alpha. 

Let us demonstrate that different training and testing datasets result in trees with different accuracies when we set ccp_alpha=0.016 using 5-fold cross validation.

```python
clf_dt=DecisionTreeClassifier(random_state=42,ccp_alpha=0.016)

scores=cross_val_score(clf_dt,X_train,y_train,cv=5)
df=pd.DataFrame(data={'tree':range(5),'accuracy':scores})
df.plot(x='tree',y='accuracy',marker='o',linestyle='--')
```

## Part-2: Coss validation to get ideal value for alpha.


```python
alpha_loop_values=[]
#to store the result of each fold during cross validation

for ccp_alpha in ccp_alphas:
    clf_dt=DecisionTreeClassifier(random_state=0,ccp_alpha=ccp_alpha)
    scores=cross_val_score(clf_dt,X_train,y_train,cv=5)
    alpha_loop_values.append([ccp_alpha,np.mean(scores),np.std(scores)])

alpha_results=pd.DataFrame(alpha_loop_values,columns=['alpha','mean_accuracy','std'])

alpha_results.plot(x='alpha',y='mean_accuracy',yerr='std',marker='o',linestyle='--')
```

Using cross validation we can see that, over all, instead of ccp_alpha=0.016, we need to set it to something closer to 0.014. We can find the exact value with:

```python
ideal_ccp_alpha=alpha_results[(alpha_results['alpha']>0.014) & (alpha_results['alpha']<0.015)]['alpha']
#ideal_ccp_alpha is currently a series
ideal_ccp_alpha=float(ideal_ccp_alpha.iloc[0])
ideal_ccp_alpha
```

## Building the Final Classification Tree

```python
clf_dt_pruned=DecisionTreeClassifier(random_state=42,ccp_alpha=ideal_ccp_alpha)
clf_dt_pruned=clf_dt_pruned.fit(X_train,y_train)
```

```python
ConfusionMatrixDisplay.from_estimator(clf_dt_pruned,X_test,y_test,display_labels=['No HD','Yes HD'])
```

In the Confusio matrix, it is seen that of the 34+8=42 people that did not have Heart Disease, 34 (81%) were correctly classified. And of the 5+28=33 people that have Heart Disease, 28 (85%) were correctly classified. There is improvement in comparison to the full-sized tree.


```python
plt.figure(figsize=(15,7.5))
plot_tree(clf_dt_pruned,filled=True,rounded=True,class_names=['No HD','YES HD'],feature_names=X_encoded.columns)
          
```
